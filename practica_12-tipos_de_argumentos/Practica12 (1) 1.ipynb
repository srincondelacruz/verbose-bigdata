Crea una función encontrar_maximo que use *args para recibir varios números y retorne el mayor de todos. No uses la función max() de Python, implementa tu propia lógica.

```python
def encontrar_maximo(*args):
    if not args:
        print("Se debe proporcionar al menos un número")
        return None
    maximo = args[0]
    for numero in args[1:]:
        if numero > maximo:
            maximo = numero
    return maximo
```

```python
resultado = encontrar_maximo(9,3,4,5,10,18,30,20,1)
print(resultado)
```

Define una función mostrar_datos_producto que use **kwargs para recibir información sobre un producto (nombre, precio, stock, categoría, etc.) e imprima cada dato en una línea separada con formato: "clave: valor".

```python
def mostrar_datos_producto(**kwargs):
    if not kwargs:
        print("No se proporcionaron datos del producto")
        return
    for clave, valor in kwargs.items():
        print(clave + ": " + str(valor))
```

```python
mostrar_datos_producto(nombre="portatil", precio=1200,stock=5, categoria="informatica", calidad="ultra")
```

Escribe una función `procesar_pedido` que reciba:

- Un argumento posicional: `numero_pedido`
- Argumentos posicionales variables: `productos` (nombres de productos)
- Dos argumentos de palabra clave obligatorios: `cliente` y `direccion`
- Argumentos de palabra clave variables: `*opciones_envio`

La función debe imprimir toda la información del pedido de forma organizada.

```python
def procesar_pedido(numero_pedido, *productos, cliente, direccion, **opciones_envio):
    print("=== Información del Pedido ===")
    print(f"Número de pedido: {numero_pedido}")
    print(f"Cliente: {cliente}")
    print(f"Dirección: {direccion}")

    print("Productos:")
    if productos:
        for i, producto in enumerate(productos, 1):
            print(f"  {i}. {producto}")
    else:
        print("  (No se especificaron productos)")

    print("Opciones de envío:")
    if opciones_envio:
        for clave, valor in opciones_envio.items():
            print(f"  {clave}: {valor}")
    else:
        print("  (No se especificaron opciones de envío)")
    print("=============================")
```

```python
procesar_pedido(
    101, 
    "Libro", "Boli",
    cliente="Lucia",
    direccion="Calle Falsa 123",
    entrega="urgente", 
    horario="mañana"
)
procesar_pedido(
    102,
    cliente="javi",
    direccion="Avenida libertad 123"
)
```

Crea una función calcular_estadisticas que reciba números como *args y retorne un diccionario con: suma, promedio, mínimo y máximo de esos números.

```python
def calcular_estadisticas(*numeros):
    if not numeros:
        return {"suma": 0, "promedio": 0, "mínimo": None, "máximo": None}
    suma = sum(numeros)
    promedio = suma / len(numeros)
    minimo = min(numeros)
    maximo = max(numeros)
    return {
        "suma": suma,
        "promedio": promedio,
        "mínimo": minimo,
        "máximo": maximo
    }
```

```python
resultado = calcular_estadisticas(5, 10, 2, 8)
print(resultado)
resultado_vacio = calcular_estadisticas()
print(resultado_vacio)
```

Crea un programa con:

- Una variable global `saldo = 1000`
- Una función `depositar(cantidad)` que añada dinero al saldo
- Una función `retirar(cantidad)` que reste dinero del saldo
- Una función `mostrar_saldo()` que imprima el saldo actual

Asegúrate de usar correctamente la palabra clave `global` donde sea necesario.

```python
saldo = 1000

def depositar(cantidad):
    global saldo
    saldo += cantidad

def retirar(cantidad):
    global saldo
    saldo -= cantidad

def mostrar_saldo():
    print(f"El saldo actual es: {saldo}")
```

```python
mostrar_saldo()
depositar(500)
mostrar_saldo()
retirar(300)
mostrar_saldo()
```
